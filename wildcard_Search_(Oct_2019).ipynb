### Kommando for å lete etter stavelsesvarianter med jokertegn.
Utviklet av Lars G. Johnsen. 


```python
import pandas as pd
import requests
```

```python
def wildcardsearch(params = {'word': '', 'freq_lim':50, 'limit':50, 'factor':2}):
    res = requests.get('https://api.nb.no/ngram/wildcards', params = params)
    if res.status_code == 200:
        result = res.json()
    else:
        result = {'status':'feil'}
    resultat = pd.DataFrame.from_dict(result, orient='index')
    if not(resultat.empty):
        resultat.columns = [params['word']]
    return resultat

def sorted_wildcardsearch(params):
    res = wildcardsearch(params)
    if not res.empty:
        res = res.sort_values(by=params['word'], ascending=False)
    return res
```

### Legg inn ordet det søkes etter med * for åpne plasser.

Sånn sett er det et vanlig søk med * som jokertegn. Parameteret factor bestemmer størrelsen på søket, og gir en slags Levenshtein-effekt. For å se på sammensatte ord økes factor til å få med det. Søket må startes med en eller flere bokstaver, jo fler bokstaver i starten, desto raskere går søket.


```python
sorted_wildcardsearch(
    {
        'word': 'Kj*rligh*d',   # her legges selve søkeordet inn
        'factor':1,     # factor bestemmer hvor mye lenger treffene skal være enn ordet med jokertegn.
        'freq_lim':1, # sett begrensninger på frekvensen, minimumsverdi
        'limit':200    # begrensning på antall treff
        
    })
```
